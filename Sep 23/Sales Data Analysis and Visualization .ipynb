```python
import pandas as pd
import matplotlib.pyplot as plt
# Sales dataset
data = {
"Month": ["Jan", "Feb", "Mar", "Jan", "Feb", "Mar", "Jan", "Feb", "Mar"],
"Product": ["Mobile", "Mobile", "Mobile", "Laptop", "Laptop", "Laptop", "TV",
"TV", "TV"],
"Units_Sold": [200, 220, 210, 120, 135, 150, 80, 95, 100],
"Revenue": [300000, 330000, 315000, 600000, 675000, 750000, 240000, 285000,
300000]
}
df = pd.DataFrame(data)
df
```

```python
#Basic Analysis 
total_revenue = df.groupby("Product")["Revenue"].sum()
avg_units_sold = df.groupby("Product")["Units_Sold"].mean() 
print(total_revenue) 
print(avg_units_sold)
```

```python
#Line Chart 
df.pivot(index = "Month", columns = "Product", values = "Revenue").plot() 
plt.title("Monthly Revenue Trend") 
plt.xlabel("Month")
plt.ylabel("Revenue")
plt.show()
```

```python
#Bar chart 
plt.figure(figsize=(6,4)) 
total_revenue.plot(kind = "bar", color=['blue', 'green', 'orange']) 
plt.title("Total Revenue per product") 
plt.xlabel("Product")
plt.ylabel("Total Revenue")
plt.show()
```

```python
#Pie chart 
total_revenue.plot(kind="pie") 
plt.title("Revenue Contribution by product") 
plt.ylabel(" ") 
plt.show()
```
